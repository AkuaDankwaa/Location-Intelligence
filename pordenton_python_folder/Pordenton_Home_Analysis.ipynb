```python
parks_gdf = geopandas.read_file('pordenton_parks/Parks.shp')
display(parks_gdf)
```

```python
parks_gdf.info()
parks_gdf.crs
```

```python
#filtering residence_gdf by 5 conditions
#Type: Single family
#sale_lease: sale
#Askprice: >= $175000 but <= $225,000
#no. of bedrooms: >=3
#build_year: >=1990

filtered_residence_gdf = residence_gdf[
    (residence_gdf['Type'] == 'Single Family House') & 
    (residence_gdf['Sale_Lease'] == 'Sale') & 
    (residence_gdf['AskPrice'] >= 175000) & 
    (residence_gdf['AskPrice'] <= 225000) & 
    (residence_gdf['Bedrooms'] >= 3) & 
    (residence_gdf['Subdivis_3'] >= 1990) 
    
    ]

#rename the 'subdivis_3' column to 'Build year'
filtered_residence_gdf.rename(columns={'Subdivis_3': 'Build_year'}, inplace=True)

display(filtered_residence_gdf)
filtered_residence_gdf.shape
    
```

```python
#filtering schools

filtered_schools_gdf = schools_gdf[schools_gdf['School'].isin(['Arnie Primary', 'Melvin Elementary'])]

#print results
display(filtered_schools_gdf)
```

```python
#merging residence and schools

#selecting columns to keep after filter
selected_columns = ['Type', 'Sale_Lease', 'AskPrice', 'Bedrooms', 'Build_year', 'geometry']

#keep only selected columns
filtered_residence_gdf = filtered_residence_gdf[selected_columns]

residence_schools_merged_gdf = geopandas.sjoin(filtered_residence_gdf, filtered_schools_gdf, how='inner', op='within')

#print result
display(residence_schools_merged_gdf)

```

```python
residence_schools_merged_gdf.shape
```

```python
#Final criterion--keep residence 3miles near Terry Athletic park
filtered_parks = parks_gdf[parks_gdf['PARK'] == 'Terry Athletic Park']

#print results
display(filtered_parks)

# mile to meters
buffer_distance =  1609.34 

projected_crs = "EPSG:3395"

#reprojecting the gdfs to same crs; creating buffer around Terry Atlantic Park
park_buffer = filtered_parks.to_crs(projected_crs).buffer(buffer_distance)
residence_schools_merged_gdf = residence_schools_merged_gdf.to_crs(projected_crs)

#convert the buffer Geoseries to a gdf
park_buffer_gdf = geopandas.GeoDataFrame(geometry=park_buffer)#, crs=park_buffer)
#park_buffer = filtered_parks.to_crs(residence_schools_merged_gdf.crs)
```

```python
#final refine selection
# rename or drop columns in residence_schools_merged_gdf
if 'index_left' in residence_schools_merged_gdf.columns or 'index_right' in residence_schools_merged_gdf.columns:
    residence_schools_merged_gdf = residence_schools_merged_gdf.rename(columns={'index_left': 'res_index_left', 'index_right': 'res_index_right'})


residence_schools_park = geopandas.sjoin(residence_schools_merged_gdf, park_buffer_gdf, predicate = 'within', how = 'inner')

#print final results
display(residence_schools_park)
```
